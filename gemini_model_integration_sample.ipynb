```python
# Prompting the gemini model
```

```python
model=genai.GenerativeModel( model_name="gemini-2.5-flash")
```

```python
user_prompt="""
Generate some factual information to complete the following sentence in 5-6 lines.
In our solar system, Earth is a ...
"""
```

```python
response=model.generate_content(user_prompt)
```

```python
print(response.text)
```

```python
# Custom Configurations 
```

```python
model=genai.GenerativeModel( model_name="gemini-2.5-flash")

custom_config=genai.types.GenerationConfig(max_output_tokens=1024,temperature=1.0)

user_prompt="""
Generate some factual information to complete the following sentence in 5-6 lines.
In our solar system, Earth is a ...
"""

response=model.generate_content(user_prompt,generation_config=custom_config)

print(response.text)
```
